```python
from matplotlib.colors import ListedColormap
from matplotlib import pyplot as plt
from sklearn import tree
import numpy as np
```

# Decision tree boundries plotting funciton 

```python
# Plot decision tree boundries - inspired by: https://github.com/ageron/handson-ml3
def plot_decision_boundary(clf, X, y, axes=[0, 7.5, 0, 3], legend=False, plot_training=True):
    x1s = np.linspace(axes[0], axes[1], 100)
    x2s = np.linspace(axes[2], axes[3], 100)
    x1, x2 = np.meshgrid(x1s, x2s)
    X_new = np.c_[x1.ravel(), x2.ravel()]
    y_pred = clf.predict(X_new).reshape(x1.shape)
    custom_cmap = ListedColormap(['#fafab0','#9898ff','#a0faa0'])
    plt.contourf(x1, x2, y_pred, alpha=0.3, cmap=custom_cmap)

    custom_cmap2 = ListedColormap(['#7d7d58','#4c4c7f','#507d50'])
    plt.contour(x1, x2, y_pred, cmap=custom_cmap2, alpha=0.8)
    if plot_training:
        plt.plot(X[:, 0][y==0], X[:, 1][y==0], "yo", label="Iris setosa")
        plt.plot(X[:, 0][y==1], X[:, 1][y==1], "bs", label="Iris versicolor")
        plt.plot(X[:, 0][y==2], X[:, 1][y==2], "g^", label="Iris virginica")
        plt.axis(axes)
    else:
        plt.xlabel(r"$x_1$", fontsize=18)
        plt.ylabel(r"$x_2$", fontsize=18, rotation=0)
    if legend:
        plt.legend(loc="lower right", fontsize=14)
```

# Example 1



Source: https://www.scientificamerican.com/article/dissect-a-flower/

```python
from sklearn.datasets import load_iris

iris = load_iris()
X = iris.data[:, 2:] # petal length and width
y = iris.target
```

```python
X.shape, y.shape
```

```python
# number of classes
np.unique(y).shape
```

```python
tree_clf = tree.DecisionTreeClassifier(max_depth=2, random_state=42)
tree_clf.fit(X, y)
```

```python
tree.plot_tree(tree_clf)
```

1. **Node Condition (e.g., `x[0] <= 2.45`)**:
   - This represents the splitting condition at the node. For example, `x[0] <= 2.45` means the decision tree splits the data based on whether the first feature (petal length in this case) is less than or equal to 2.45.

2. **Gini Index (e.g., `gini = 0.667`)**:
   - The Gini index measures the impurity of the node. A Gini index of 0 means the node is pure (all samples belong to one class), while higher values indicate more impurity. For example:
     - `gini = 0.0` means all samples in the node belong to the same class.
     - `gini = 0.667` indicates a mix of classes.

3. **Samples (e.g., `samples = 150`)**:
   - This shows the number of samples (data points) present in the node. For example, the root node has `samples = 150`, meaning all 150 data points are considered at the start.

4. **Value (e.g., `value = [50, 50, 50]`)**:
   - This represents the distribution of samples across the classes in the node. For example:
     - `value = [50, 50, 50]` means there are 50 samples from each of the three classes (Iris setosa, Iris versicolor, and Iris virginica) in the node.
     - `value = [50, 0, 0]` means all 50 samples in the node belong to the first class (Iris setosa).

5. **True/False Branches**:
   - The tree splits into two branches:
     - **True**: The condition at the node is satisfied (e.g., `x[0] <= 2.45` is true).
     - **False**: The condition at the node is not satisfied (e.g., `x[0] > 2.45`).

### Example Interpretation:
- **Root Node**:
  - Condition: `x[0] <= 2.45`
  - Gini: `0.667` (high impurity, as all three classes are present)
  - Samples: `150` (all data points)
  - Value: `[50, 50, 50]` (equal distribution of classes)

- **Left Child Node**:
  - Condition satisfied (`True` branch): `x[0] <= 2.45`
  - Gini: `0.0` (pure node, all samples belong to one class)
  - Samples: `50`
  - Value: `[50, 0, 0]` (all samples are Iris setosa)

- **Right Child Node**:
  - Condition not satisfied (`False` branch): `x[0] > 2.45`
  - Gini: `0.5` (impurity, mix of two classes)
  - Samples: `100`
  - Value: `[0, 50, 50]` (equal distribution of the remaining two classes)

This visualization helps understand how the decision tree splits the data at each step to classify it into different classes.

```python
plot_decision_boundary(tree_clf, X, y)
plt.text(1.40, 1.0, "Depth=0", fontsize=15)
plt.text(3.2, 1.80, "Depth=1", fontsize=13)
```

```python
tree_clf.get_depth()
```

```python
for name, score in zip(["length", "wedith"], tree_clf.feature_importances_):
    print(name, score)
```

# Decision tree overfitting

```python
from sklearn.datasets import make_moons
```

```python
Xm, ym = make_moons(n_samples=100, noise=0.25, random_state=53)
```

```python
deep_tree_clf1 = tree.DecisionTreeClassifier(random_state=42)
deep_tree_clf2 = tree.DecisionTreeClassifier(min_samples_leaf=4, random_state=42)
deep_tree_clf1.fit(Xm, ym)
deep_tree_clf2.fit(Xm, ym)
```

```python
# plot unrestrcited tree
plot_decision_boundary(deep_tree_clf1, Xm, ym, axes=[-1.5, 2.4, -1, 1.5])
```

```python
# restrcited (controled) tree
plot_decision_boundary(deep_tree_clf2, Xm, ym, axes=[-1.5, 2.4, -1, 1.5])
```

```python
deep_tree_clf1.get_depth(), deep_tree_clf2.get_depth()
```

```python
tree.plot_tree(deep_tree_clf1)
```
